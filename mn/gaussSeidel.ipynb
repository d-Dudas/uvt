Pentru Gauss-Siedel se va folosi formula de pe slide-urile cursului



```python
import numpy as np

def gaussSiedel(a, b, maxNumberOfIterations, errorTolerance):
    n = len(a)
    x = np.zeros(n)

    for k in range(maxNumberOfIterations):
        x_old = np.copy(x)

        for i in range(n):
            s1 = np.dot(a[i, :i], x[:i])
            s2 = np.dot(a[i, i+1:], x[i+1:])
            x[i] = (b[i] - s1 - s2) / a[i, i]

        print("k: {}, x: {}".format(k, x))

        if np.max(np.abs(x - x_old)) < errorTolerance:
            break
    
    return x
```

```python
# Exercitiul exemplu din curs

a = np.array([[5, -2, 3], [-3, 9, 1], [2, -1, -7]])
b = np.array([-1, 2, 3])

maxNumberOfIterations = 100
errorTolerance = 1e-3

print(gaussSiedel(a, b, maxNumberOfIterations, errorTolerance))
```
