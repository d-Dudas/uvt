```python
from numpy.typing import NDArray

def s_liniar(x_data: NDArray, y_data: NDArray, val: float, i: int)-> float:
    return y_data[i] + ((y_data[i] - y_data[i-1]) / (x_data[i] - x_data[i-1])) * (val - x_data[i])
```

```python
from numpy.typing import NDArray

def spline_liniar(x_data: NDArray, y_data: NDArray, val: float):
    for i in range(1, len(x_data)):
        if x_data[i-1] <= val and x_data[i] > val:
            return s_liniar(x_data, y_data, val, i)
```

```python
import numpy
from math import pi
from math import sin
from numpy.typing import NDArray
import matplotlib.pyplot as plt

x_data: NDArray = numpy.array([0, pi/2, pi, (3*pi)/2, 2*pi], dtype=float)
y_data: NDArray = numpy.array([0, 1, 0, -1, 0])

val_range = numpy.linspace(0, 2*pi)

sinus_vals = [sin(x) for x in val_range]
spline_liniar_vals = [spline_liniar(x_data, y_data, x) for x in val_range]

plt.figure(figsize=(10, 6))

plt.plot(val_range, sinus_vals, label='Sinus')
plt.plot(val_range, spline_liniar_vals, label='Spline')
plt.scatter(x_data, y_data, color='red', label='Data points')
plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)
plt.show()
```
