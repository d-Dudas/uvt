Pentru Gauss-Jacobi se foloseste formula de pe slide:



```python
import numpy as np

# Gauss-Jacobi
def gaussJacobi(a, b, maxNumberOfIterations, errorTolerance):
    n = len(a)

    # Setam valorile initiale
    x0 = np.zeros(n)
    x1 = np.zeros_like(x0)

    iterationCount = 0

    while iterationCount < maxNumberOfIterations:
        iterationCount += 1

        for i in range(n):
            # Deoarece in formula suma este pentru j=1, j!=i
            # din np.sum o sa scadem a[i, i]
            sum = np.sum(a[i, :] * x0) - a[i, i] * x0[i]
            x1[i] = (b[i] - sum) / a[i, i]
        
        # Daca cea mai mare diferenta de valori dintre
        # rezultate obitnute in iteratia curenta
        # este mai mica decat toleranta de eroare,
        # atunci ne putem opri
        if np.max(np.abs(x1 - x0)) < errorTolerance:
            break

        x0 = x1.copy()
    
    return x1, iterationCount
```

```python
# Un exercitiu de pe slide-urile de pe curs

a = np.array([[5, -2, 3], [-3, 9, 1], [2, -1, -7]])
b = np.array([-1, 2, 3])

errorTolerance = 1e-3
maxNumberOfIterations = 3

print(gaussJacobi(a, b, maxNumberOfIterations, errorTolerance))
```
